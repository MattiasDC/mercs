# Expanded Induction

Learn ensembles, expand those trees to MERCS as a whole.

## Preliminaries

```python
import mercs
import numpy as np
from mercs.tests import load_iris, default_dataset
from mercs.core import Mercs
```

## Setup

```python
train, test = default_dataset(n_features=4)
```

# Sandbox

```python
clf = Mercs(
    max_depth=3,
    classifier_algorithm="XGB",
    regressor_algorithm="XGB",
    selection_algorithm="random",
    induction_algorithm="expand",
    fraction_missing=0.,
    nb_targets=1,
    nb_iterations=1,
    n_jobs=1,
    verbose=0,
    inference_algorithm="own",
    n_estimators=30,
    max_steps=8,
    prediction_algorithm="rw",
    random_state=800,
)
```

```python
clf.fit(train)
```
